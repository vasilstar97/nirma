```python
from langchain_community.tools import DuckDuckGoSearchResults

# 1. Инициализируем инструмент поиска
search_tool = DuckDuckGoSearchResults(
    name="web_search",
    output_format='list'
)
```

```python
from langchain.agents import create_agent
from src import llm

agent = create_agent(model=llm, tools=[search_tool], debug=True, system_prompt='Ты эксперт по стратегическому и территориальному планированию.')
```

```python
result = agent.invoke({"messages": "Разработай целеполагание для г. Тюмень"})
```
